```python
# handling the null values. In this dataset 'Not Available', 'unknown' are null values.

df['Ship Mode'].unique()
```

```python
df = pd.read_csv('orders.csv', na_values=['Not Available', 'unknown'])
```

```python
df['Ship Mode'].unique()
```

```python
# renaming column names to lowercase and replacing space with underscores

df.columns = df.columns.str.lower()
df.columns = df.columns.str.replace(' ','_')
```

```python
df.columns
```

```python
# derive new columns discount, sales price and profit
```

```python
df['discount']=df['list_price']*df['discount_percent']*0.01
```

```python
df['sale_price']=df['list_price']-df['discount']
```

```python
df['profit']=df['sale_price']-df['cost_price']
```

```python
df.head()
```

```python
# convert date datatype to datetime
```

```python
df['order_date'] = pd.to_datetime(df['order_date'], format = '%Y-%m-%d')
```

```python
# drop unneccessary columns cost_price, list_price, discount_percent
```

```python
df.drop(columns = ['list_price','cost_price','discount_percent'], inplace=True)
```
